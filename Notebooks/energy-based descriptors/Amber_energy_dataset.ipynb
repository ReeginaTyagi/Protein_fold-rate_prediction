```python
import os
import csv

# Directory path
directory = r"C:\Users\sptrp\OneDrive\Documents\Reegina\mine\pri"

# Initialize a dictionary to store counts
protein_counts = {}

# Traverse through the directory
for root, dirs, files in os.walk(directory):
    # Check if the current directory is a saltbridges directory
    if os.path.basename(root) == 'saltbridges':
        # Extract the protein name from the parent directory
        protein_name = os.path.basename(os.path.dirname(root))

        # Count the number of .dat files
        dat_files_count = len([file for file in files if file.endswith('.dat')])

        # Update the count in the dictionary
        if protein_name in protein_counts:
            protein_counts[protein_name] += dat_files_count
        else:
            protein_counts[protein_name] = dat_files_count

# Write the counts to a CSV file
csv_file_path = os.path.join(directory, 'protein_counts.csv')
with open(csv_file_path, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Protein Name', 'File Count'])
    for protein, count in protein_counts.items():
        writer.writerow([protein, count])

print("CSV file has been created with the counts.")
```

```python
import pandas as pd
import os
import re

def calculate_statistics(folder_path):
    statistics_values = []
    protein_folder_pattern = re.compile(r"\d+[A-Z]+")  # Pattern to match folder names like '1A6N', '4BLM'
    for root, dirs, files in os.walk(folder_path):
        # Check if the current directory name matches the pattern of a protein folder
        if re.search(protein_folder_pattern, os.path.basename(root)):
            for file in files:
                if file == "rmsf.dat":
                    file_path = os.path.join(root, file)
                    try:
                        df = pd.read_csv(file_path, delim_whitespace=True, header=None)
                        if len(df.columns) >= 2:  # Check if there is at least two columns
                            min_value = df[1].min()  # Compute minimum of the second column
                            max_value = df[1].max()  # Compute maximum of the second column
                            average_value = df[1].mean()  # Compute average of the second column
                            statistics_values.append({
                                "Protein": os.path.basename(root),
                                "Min_Value": min_value,
                                "Max_Value": max_value,
                                "Average_Value": average_value
                            })
                    except Exception as e:
                        print(f"Failed to process {file_path}: {e}")

    return statistics_values

def main():
    parent_directory = r"C:\Users\sptrp\OneDrive\Documents\Reegina\mine\pri"
    stats = calculate_statistics(parent_directory)
    if stats:
        # Convert list of dictionaries to DataFrame
        df_stats = pd.DataFrame(stats)
        # Specify the file path for the CSV output
        output_file_path = os.path.join(parent_directory, "rmsf_statistic.csv")
        # Write DataFrame to a CSV file
        df_stats.to_csv(output_file_path, index=False)
        print(f"Statistics saved to {output_file_path}")
    else:
        print("No data found or processed.")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR

# Load the datasets
energy_df = pd.read_csv('/content/drive/MyDrive/Reegina /Data/energy-based descriptor/energy_all_proteins.csv', encoding = 'latin1')
energy_2s_df = pd.read_csv('/content/drive/MyDrive/Reegina /Data/energy-based descriptor/energy_2s.csv', encoding = 'latin1')
energy_52_df = pd.read_csv('/content/drive/MyDrive/Reegina /Data/energy-based descriptor/energy_2s_single_domain.csv', encoding = 'latin1')
```

```python
# Data preparation
def prepare_data(df):
    if 'PROTEIN IDS' in df.columns:
        df = df.drop(['PROTEIN IDS'], axis=1)
    elif 'PDB' in df.columns:
        df = df.drop(['PDB'], axis=1)
    else:
        raise ValueError("Neither 'PROTEIN IDS' nor 'PDB' column found in the DataFrame")

    X = df.drop(['ln(kf)'], axis=1)
    y = df['ln(kf)']

    X_scaled = StandardScaler().fit_transform(X)

    return train_test_split(X_scaled, y, test_size=0.3, random_state=2)

X_train_energy, X_test_energy, y_train_energy, y_test_energy = prepare_data(energy_df)
X_train_2s, X_test_2s, y_train_2s, y_test_2s = prepare_data(energy_2s_df)
X_train_52, X_test_52, y_train_52, y_test_52 = prepare_data(energy_52_df)

# Gradient Boosting Regression
def gradient_boosting(X_train, y_train, X_test, y_test):
    model = GradientBoostingRegressor(n_estimators=100, random_state=1)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, r2

# Support Vector Regression with parameter tuning
def support_vector_regression(X_train, y_train, X_test, y_test):
    param_grid = {'C': [0.1, 1, 10, 100], 'gamma': ['scale', 'auto', 0.01, 0.1, 1]}
    grid_search = GridSearchCV(SVR(kernel='rbf'), param_grid, cv=5, scoring='neg_mean_squared_error', verbose=2)
    grid_search.fit(X_train, y_train)
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return grid_search.best_params_, mse, r2

# Run models
gb_mse_energy, gb_r2_energy = gradient_boosting(X_train_energy, y_train_energy, X_test_energy, y_test_energy)
svr_params_energy, svr_mse_energy, svr_r2_energy = support_vector_regression(X_train_energy, y_train_energy, X_test_energy, y_test_energy)

# Outputs (to be printed or logged as required)
print("Gradient Boosting MSE:", gb_mse_energy, "R2:", gb_r2_energy)
print("SVR Best Params:", svr_params_energy, "MSE:", svr_mse_energy, "R2:", svr_r2_energy)
```
