```python
import os
import numpy as np
from Bio.PDB import PDBParser
from scipy.spatial import distance

# Dictionary to map Bio.PDB residue names to amino acid symbols
amino_lookup = {
    'ALA': 'A', 'CYS': 'C', 'ASP': 'D', 'GLU': 'E', 'PHE': 'F',
    'GLY': 'G', 'HIS': 'H', 'ILE': 'I', 'LYS': 'K', 'LEU': 'L',
    'MET': 'M', 'ASN': 'N', 'PRO': 'P', 'GLN': 'Q', 'ARG': 'R',
    'SER': 'S', 'THR': 'T', 'VAL': 'V', 'TRP': 'W', 'TYR': 'Y'
}

def crawl_pdb(base_dir):
    """
    This function reads PDB files from separate folders named after their IDs
    within the specified base directory, extracts coordinates, distance matrices,
    and sequences for each PDB structure. It saves the distance matrix as a CSV
    and the sequence as a text file.

    Args:
    - base_dir (str): Path to the base directory containing folders named after PDB IDs.

    Returns:
    - pdbinfo_dict (dict): Dictionary containing PDB information for each file.
                          Keys are PDB file names, values are dictionaries with keys:
                            - 'coords': Matrix of atomic coordinates (Nx3).
                            - 'distance_matrix': Distance matrix (NxN).
                            - 'sequence': Amino acid sequence string.
    """
    parser = PDBParser(QUIET=True)
    pdbinfo_dict = {}

    # List all folders (each folder corresponds to a PDB ID)
    folders = [folder for folder in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, folder))]

    for pdb_id in folders:
        pdb_path = os.path.join(base_dir, pdb_id)

        # Assuming each folder contains only one PDB file named after the folder
        pdb_file = os.path.join(pdb_path, f'{pdb_id}.pdb')

        if not os.path.exists(pdb_file):
            continue

        info = {}
        info['id'] = pdb_id
        structure = parser.get_structure(pdb_id, pdb_file)

        # Extract coordinates and labels
        coordinates = []
        labels = []

        for model in structure:
            for chain in model:
                for residue in chain:
                    residue_name = residue.get_resname()
                    if residue_name not in ['HOH', ' CA']:  # Exclude water and CA residues
                        try:
                            coordinates.append(residue['CA'].get_coord())
                            labels.append(residue_name)
                        except KeyError:
                            pass

        coords = np.asarray(coordinates)
        if len(coords) < 2:
            continue  # Skip structures with fewer than 2 residues

        # Compute distance matrix
        distance_matrix = distance.squareform(distance.pdist(coords))

        # Save distance matrix as CSV
        distance_csv_file = os.path.join(pdb_path, 'distance_matrix.csv')
        np.savetxt(distance_csv_file, distance_matrix, delimiter=',')

        # Generate sequence string
        sequence = ''.join([amino_lookup.get(s, 'X') for s in labels])  # 'X' for unknown residues

        # Save sequence as text file
        sequence_txt_file = os.path.join(pdb_path, 'sequence.txt')
        with open(sequence_txt_file, 'w') as seq_file:
            seq_file.write(sequence)

        # Store information in dictionary
        info['coords'] = coords
        info['distance_matrix'] = distance_matrix
        info['sequence'] = sequence

        pdbinfo_dict[pdb_id] = info

    return pdbinfo_dict

# Example usage:
base_directory = '/content/drive/MyDrive/Reegina /Data/all_pdb with distance matrix'
pdb_data = crawl_pdb(base_directory)
```
