```python
import numpy as np
property_names = ['K°', 'Ht', 'Hp', 'P', 'pHi', 'pK`', 'Mw', 'Bl', 'Rf', 'μ', 'Hnc', 'Esm', 'El', 'Et', 'Pα', 'Pβ', 'Pt', 'Pc', 'Ca', 'F', 'Pr', 'Ra', 'Ns', 'αn', 'αc', 'αm', 'V°', 'Nm', 'Nl', 'Hgm', 'ASAD', 'ASAN', 'ΔASA', 'ΔGh', 'GhD', 'GhN', 'ΔHh', '-TΔSh', 'ΔCph', 'ΔGc', 'ΔHc', '-TΔSc', 'ΔG', 'ΔH', '-TΔS', 'V', 's', 'f']

# Function to calculate average properties
def average_properties(sequence, properties_dict):
    sum_properties = np.zeros(len(property_names))
    valid_counts = 0

    for amino_acid in sequence:
        if amino_acid in properties_dict:
            sum_properties += np.array([properties_dict[amino_acid][prop] for prop in property_names])
            valid_counts += 1

    return sum_properties / valid_counts if valid_counts > 0 else np.full(len(property_names), np.nan)

# Applying the function to each sequence in the DataFrame
protein_df['AverageProperties'] = protein_df['sequences'].apply(lambda x: average_properties(x, properties_normalized))

# Expanding the average properties into separate columns
average_properties_df = pd.DataFrame(protein_df['AverageProperties'].tolist(), columns=property_names)

# Concatenating the average properties columns back to the original protein_df
protein_new_df = pd.concat([protein_df.drop('AverageProperties', axis=1), average_properties_df], axis=1)

# Display the final DataFrame
protein_new_df.head()
```
